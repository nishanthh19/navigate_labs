```python
from crewai import Agent, Task, Crew
from langchain.chat_models import ChatOpenAI
import os

os.environ["OPENAI_API_BASE"] = "https://api.nexus.navigatelabsai.com/v1"
os.environ["OPENAI_API_KEY"] = "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"
llm_agent = ChatOpenAI(
    model="gpt-3.5",
    temperature=0.1
)


# Define Agents (same as before)...

# Define Tasks with expected_output

# 1. Research Agent
research_agent = Agent(
    role="Market Researcher",
    goal="Gather accurate and up-to-date information about the Electric Vehicle market",
    backstory="You are skilled in gathering and synthesizing market data from various online and offline sources.",
    verbose=True,
    llm=llm
)

# 2. Analysis Agent
analysis_agent = Agent(
    role="Data Analyst",
    goal="Analyze electric vehicle market data to uncover key trends and insights",
    backstory="You are a data-savvy analyst with experience in extracting actionable insights from raw market data.",
    verbose=True,
    llm=llm
)

# 3. Writer Agent
writer_agent = Agent(
    role="Technical Writer",
    goal="Create a comprehensive and professional market research report",
    backstory="You write concise and structured technical documents for business stakeholders.",
    verbose=True,
    llm=llm
)

# 4. Reviewer Agent
reviewer_agent = Agent(
    role="Editor",
    goal="Proofread and refine the final report to ensure clarity and correctness",
    backstory="You are an expert in grammar, clarity, and tone optimization.",
    verbose=True,
    llm=llm
)


research_task = Task(
    description="Research the current state of the electric vehicle market, key players, market size, and emerging trends.",
    agent=research_agent,
    expected_output="A summary of key EV market players, market size, and trends."
)

analysis_task = Task(
    description="Analyze the research findings and identify growth patterns, customer preferences, and forecast trends.",
    agent=analysis_agent,
    context=[research_task],
    expected_output="List of major trends, insights, and growth forecasts for the EV market."
)

write_report_task = Task(
    description="Draft a professional market research report using the insights from the analysis.",
    agent=writer_agent,
    context=[analysis_task],
    expected_output="A full, well-structured market research report with sections and clear formatting."
)

review_task = Task(
    description="Review the report for accuracy, structure, and language. Suggest and apply improvements.",
    agent=reviewer_agent,
    context=[write_report_task],
    expected_output="A polished final version of the market report, free from errors."
)


crew = Crew(
    agents=[research_agent, analysis_agent, writer_agent, reviewer_agent],
    tasks=[research_task, analysis_task, write_report_task, review_task],
    verbose=True
)


# Run the crew and get the final result
result = crew.kickoff()
print("\n\n📄 Final Report:\n")
print(result)
```
